# Put item

```python
import datetime 
from pytz import timezone
def generate_string(merchant):
    now = datetime.datetime.now(timezone('US/Eastern')) 
    
    d = {}
    for key,val in merchant.items():
        if key in ['description', 'image_backup']:
            d[key] = {"SS":val}
        else:
            d[key] = {"S":val}
    d['insertedAtTimestamp']:now.strftime('%Y-%m-%dT%H:%M:%S-%Z')
    s = json.dumps(d)
    s = '\''+s+'\''
    return s
```

# Search 24 items of a keyword and parse items into list #

```python
import requests
import json
import datetime 
from pytz import timezone
def generate_string(merchant):
    now = datetime.datetime.now(timezone('US/Eastern')) 
    
    d = {}
    for key,val in merchant.items():
        if key in ['description', 'image_backup']:
            d[key] = {"SS":val}
        else:
            d[key] = {"S":val}
    d['insertedAtTimestamp']:now.strftime('%Y-%m-%dT%H:%M:%S-%Z')
    s = json.dumps(d)
    s = '\''+s+'\''
    return s



#
keywords = [ 'pear', 'pen','water', 'coffee', 'ham','tea','bread'
           'milk', 'cranberry','banana','blueberry','blackberry','mango','oil','salt',
          'pepper','sugar','baking sheet','tin roll','foil','syrup','honey','flower','flour','paper','pan','lemon','corn','cucumber','potato','carrot','cauliflower','broccoli', 'beans',
          'mushroom','rice', 'pasta','lettuce', 'asparagus', 'onion','cabbage','dressing','egg','butter','cheese','juice']

for keyword in keywords:
    merchandise_list = []

    url = "https://target-com-store-product-reviews-locations-data.p.rapidapi.com/product/search"

    querystring = {"store_id":"3991","keyword":keyword,"offset":"0","limit":"24","sponsored":"1","rating":"0"}

    headers = {
        'x-rapidapi-host': "target-com-store-product-reviews-locations-data.p.rapidapi.com",
        'x-rapidapi-key': "90edf87601msheef53f51133c0f9p149c70jsn62c12b7d797b"
        }

    response = requests.request("GET", url, headers=headers, params=querystring)

    y = json.loads(response.text)
    num = len(y['products']) #default 24
    print("Keyword: "+keyword+" Found " + str(num))

    for i in range(num):
        item = y['products'][i]
        tcin = item['tcin']
        if 'enrichment' in item['item'] and 'images' in item['item']['enrichment'] and 'primary_image_url' in item['item']['enrichment']['images']:
            image_url = item['item']['enrichment']['images']['primary_image_url']
        if 'alternate_image_urls' in item['item']['enrichment']['images']:
            image_backup = item['item']['enrichment']['images']['alternate_image_urls']
        if 'soft_bullets' in item['item']['product_description'] and 'bullets' in item['item']['product_description']['soft_bullets']:
            description = item['item']['product_description']['soft_bullets']['bullets']
        if 'product_description' in item['item'] and 'title' in item['item']['product_description']:
            title = item['item']['product_description']['title']
        if 'parent' in item:
            price_range = item['parent']['price']['formatted_current_price']
        if 'formatted_current_price' in item['price']:
            price = item['price']['formatted_current_price']
        merchandise = {'id':tcin, 'keyword':keyword, 'title':title, 'price':price, 'image_url':image_url, 'description':description, 'price_range':price_range, 'image_backup':image_backup}
        merchandise_list.append(merchandise)
        attribute_s = generate_string(merchandise)
        !aws dynamodb put-item --table-name items --item $attribute_s
        print("Keyword: "+keyword+" Uploaded " + str(i))

    # dump list into json file 
    with open('/Users/yumengluo/Desktop/claire/Columbia/6998/6998-team/'+keyword+'.json', 'w') as file:
        json.dump(merchandise_list, file)
```
